```python
import math
import matplotlib.pyplot as plt
from matplotlib import mlab
# Вводим начальные условия и интервал
x0 = 0
xn = 10
u0 = 1
m0 = 1
# Выбираем шаг
h = 1.0
uf = lambda u, m: 998 * u + 3996 * m
mf = lambda u,m: -499.5 * u - 1999 * m
#Вычисляем число шагов с учетом выбранного шага
N = int((xn-x0)/float(h))
print ("N =", N, ",h =", h)
#Создаем массив переменных x с учетом шага 
xlist = [(x0 + h * i) for i in range(N+1)]
print ("x_0..x_n =", xlist)
ulist = [0 for i in range(N+1)]
mlist = [0 for i in range(N+1)]
#Создаем массив для точного решения системы
listu = []
listm = []
#Задаем начальные условия
ulist[0] = u0
mlist[0] = m0
print("\n")
print("Решение явным методом Эйлера")
print ("x 0 = 0, u 0 = 1")
print ("x 0 = 0, m 0 = 1")
for n in range(N):
    ulist[n+1] = ulist[n] + h * uf(ulist[n], mlist[n])
    mlist[n+1] = mlist[n] + h * mf(ulist[n], mlist[n])
    print("x", n+1, "=", xlist[n+1], ",u", n+1, "=", ulist[n+1])
    print("x", n+1, "=", xlist[n+1], ",m", n+1, "=", mlist[n+1])
print("\n")
print("Решение неявным методом Эйлера")
print ("x 0 = 0, u 0 = 1")
print ("x 0 = 0, m 0 = 1")
for n in range(N):
    ulist[n+1] = ulist[n] + h * uf(ulist[n+1],mlist[n+1])
    mlist[n+1] = mlist[n] + h * mf(ulist[n+1],mlist[n+1])   
    print("x", n+1, "=", xlist[n+1], ",u", n+1, "=", ulist[n+1])
    print("x", n+1, "=", xlist[n+1], ",m", n+1, "=", mlist[n+1])
print("\n")
print("Точное решение системы")
for n in range(N+1):
    u = 6 * math.exp(-1 * xlist[n]) - 5 * math.exp(-1000 * xlist[n])
    listu.append(u)
    print("x", n, "=", xlist[n], ",u", n, "=", u)
print("\n")
for n in range(N+1):
    m = 2.5 * math.exp(-1000 * xlist[n]) - 1.5 * math.exp(-1 * xlist[n])
    listm.append(m)
    print("x", n, "=", xlist[n], ",m", n, "=", m)

plt.rc('font',**{'family':'verdana'})
plt.xlabel("x_t")
plt.ylabel("u(x)")
plt.plot(xlist, listu, "r", label = "точное решение u(x)")
plt.legend()
plt.grid()
plt.show()

plt.rc('font',**{'family':'verdana'})
plt.xlabel("x_t")
plt.ylabel("v(x)")
plt.plot(xlist, listm, "r", label = "точное решение m(x)")
plt.legend()
plt.grid()
plt.show()
```
